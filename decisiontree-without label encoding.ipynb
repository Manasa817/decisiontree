```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("/Users/kokate/Downloads/diabetes_csv.csv")
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn import tree
```

```python
X=df.drop(['class'],axis=1)#Features
y=df['class']#Target Variable
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=None)
```

```python
clf=DecisionTreeClassifier()

#Train Decision Tree classifier

clf=clf.fit(X_train,y_train)

#predict the response from test dataset

y_pred=clf.predict(X_test)
```

```python
#model accuracy ,how often the is the classifier correct?
print('accuracy:',metrics.accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
